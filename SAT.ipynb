```python
allMajors = stats["2019"]["Major"].tolist() 
allYears = list(stats.keys())
ChemEngNumberOfApplications = []
CompEngNumberOfApplications = []
DataSciNumberOfApplications = []
DigSysNumberOfApplications = []
QuantumNumberOfApplications = []

for year in allYears:
    ChemEngNumberOfApplications.append(stats[year]["Number of applications"][0])
    CompEngNumberOfApplications.append(stats[year]["Number of applications"][1])
    DataSciNumberOfApplications.append(stats[year]["Number of applications"][2])
    DigSysNumberOfApplications.append(stats[year]["Number of applications"][3])
    QuantumNumberOfApplications.append(stats[year]["Number of applications"][4])

# Plot the years against the number of applications

plt.figure(figsize=(8, 5))

plt.ylim([0, 1400])
plt.plot(allYears, DataSciNumberOfApplications, label="Data Science")
plt.plot(allYears, CompEngNumberOfApplications, label="Computational Engineering")
plt.plot(allYears, DigSysNumberOfApplications, label="Digital Systems and Design")
plt.plot(allYears, QuantumNumberOfApplications, label="Quantum Technology")
plt.plot(allYears, ChemEngNumberOfApplications, label="Chemical Engineering")


plt.xlabel("Year")
plt.ylabel("Number of applications")
plt.title("Number of applications for each major")
plt.legend()
```

```python
ChemEngIntake = []
CompEngIntake = []
DataSciIntake = []
DigSysIntake = []
QuantumIntake = []

for year in allYears[1:]:
    ChemEngIntake.append(stats[year]["Target intake"][0])
    CompEngIntake.append(stats[year]["Target intake"][1])
    DataSciIntake.append(stats[year]["Target intake"][2])
    DigSysIntake.append(stats[year]["Target intake"][3])
    QuantumIntake.append(stats[year]["Target intake"][4])

# Plot the years against the target intake

plt.figure(figsize=(8, 5))

plt.plot(allYears[1:], DataSciIntake, label="Data Science")
plt.plot(allYears[1:], CompEngIntake, label="Computational Engineering")
plt.plot(allYears[1:], DigSysIntake, label="Digital Systems and Design")
plt.plot(allYears[1:], QuantumIntake, label="Quantum Technology")
plt.plot(allYears[1:], ChemEngIntake, label="Chemical Engineering")

plt.ylim([0, 50])
plt.xlabel("Year")
plt.ylabel("Number of target intake")
plt.title("Number of target intakes for each major")
plt.legend()
```

```python
ChemEngSAT = []
CompEngSAT = []
DataSciSAT = []
DigSysSAT = []
QuantumSAT = []

for year in allYears:
    ChemEngSAT.append(stats[year]["Group II lowest SAT"][0])
    CompEngSAT.append(stats[year]["Group II lowest SAT"][1])
    DataSciSAT.append(stats[year]["Group II lowest SAT"][2])
    DigSysSAT.append(stats[year]["Group II lowest SAT"][3])
    QuantumSAT.append(stats[year]["Group II lowest SAT"][4])

# Plot the years against the lowest SAT score

plt.figure(figsize=(8, 5))

plt.plot(allYears, DataSciSAT, label="Data Science")
plt.plot(allYears, CompEngSAT, label="Computational Engineering")
plt.plot(allYears, DigSysSAT, label="Digital Systems and Design")
plt.plot(allYears, QuantumSAT, label="Quantum Technology")
plt.plot(allYears, ChemEngSAT, label="Chemical Engineering")

plt.ylim([1100, 1600])
plt.xlabel("Year")
plt.ylabel("Lowest SAT score")
plt.title("Lowest SAT score for each major")
plt.legend()
```

```python
def scoreOfCompetitiveness(major, year):
    numberOfApplications = stats[year]["Number of applications"][allMajors.index(major)]
    targetIntake = stats[year]["Target intake"][allMajors.index(major)]
    lowestSAT = stats[year]["Group II lowest SAT"][allMajors.index(major)]
    score = targetIntake/numberOfApplications 
    return score

# Plot the years against the competitiveness score

plt.figure(figsize=(8, 5))

plt.plot(allYears, [scoreOfCompetitiveness("Data Science", year) for year in allYears], label="Data Science")
plt.plot(allYears, [scoreOfCompetitiveness("Computational Engineering", year) for year in allYears], label="Computational Engineering")
plt.plot(allYears, [scoreOfCompetitiveness("Digital Systems and Design", year) for year in allYears], label="Digital Systems and Design")
plt.plot(allYears, [scoreOfCompetitiveness("Quantum Technology", year) for year in allYears], label="Quantum Technology")
plt.plot(allYears, [scoreOfCompetitiveness("Chemical Engineering", year) for year in allYears], label="Chemical Engineering")

plt.xlabel("Year")
plt.ylabel("Acceptance rate")
plt.title("Acceptance rate for each major")
plt.legend()

```
